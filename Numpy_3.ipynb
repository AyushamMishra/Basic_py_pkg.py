### Practice in Numpy

```python
import numpy as np
import matplotlib.pyplot as plt

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

```

### Knowing about Data

```python
print("==== Sales Data ====")
print("Sales data shape:",sales_data.shape,"\n","Sales data size:",sales_data.size,"\n","Sales data dimensions:",sales_data.ndim,"\n","Sales data type:",sales_data.dtype)
```

### Printing sales of 1 st three restaurents

```python
print("The sales data of first three restaurents is:","\n",sales_data[0:3])
```

### Total sales per year

```python
## Hard coded solution:

for i in range(0, 5):                                    ## Iterates over rows from 0 to 4 
    sales_each_year = 0                                  ## Iterates over columns from 1 to 4 
    for j in range(1, 5):
        sales_each_year += sales_data[i, j]
    print(f"Year {i} total sales = {sales_each_year}")


## Numpy Implementation:

sales_each_year=np.sum(sales_data[0:5,1:5],axis=1)
print("The sales each year is:",sales_each_year)
```

## Minimum Sales per year

```python
print("The minimum sales per year is:","\n",np.min(sales_data[:,1:5],axis=0))
```

## Maximum sales per year

```python
max_sales=np.max(sales_data[:,1:5],axis=0)
print("The maximum sales per year is:",max_sales)
```

## Average sales per year

```python
avg_sales=np.mean(sales_data[:,1:5],axis=1)
print("The average sales of :",avg_sales)
```

## Cumulative sum

```python
cumsum=np.cumsum(sales_data[:,1:5],axis=1)
print("The cumulative sum is:",cumsum)
```

## PLOTTING AVERAGE CUMULATIVE SUM FIGURE

```python
plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulariive sum")
plt.xlabel("Years")
plt.ylabel("mean")
plt.grid(True)
plt.show()
```

## Broadcast in Numpy

```python
print("Monthly average:","\n",sales_data[:,1:]/12)
```
