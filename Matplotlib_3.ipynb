```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('products-100.csv')
df
```

```python
fig,ax=plt.subplots(2,2,figsize=(15,15))
ax[0,0].plot(df['Index'],df['Price'],color='blue',marker='o',label='Price')
ax[0,0].set_title('Price of products')
ax[0,0].tick_params(axis='x',rotation=45,width=2)
plt.legend()
ax[0,1].bar(df['Index'],df['Stock'],color='red',label='Stock')
ax[0,1].set_title('Stock of products')
ax[0,1].tick_params(axis='x',rotation=45,width=2)
plt.legend()
ax[1,0].scatter(df['Index'],df['Internal ID'],color='green',label='Internal id')        
ax[1,0].set_title('Rating of products')
ax[1,0].tick_params(axis='x',rotation=45,width=2) 
plt.legend()
plt.tight_layout()
ax[1,1].hist(df['Stock'],bins=20,color='purple',label='Currency',edgecolor='black')
ax[1,1].tick_params(axis='x',rotation=45,width=5)
plt.legend()
plt.show()
```
