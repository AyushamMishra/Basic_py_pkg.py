## Join, merge and concatenate

## Concatenation

```python
import pandas as pd

sales_jan={
    'Customer':['A','B','C','D'],
    'Product':['Apple','Banana','Orange','Pineapple'],
    'Quantity':[5,3,6,8],
    'Price':[100,200,300,400]

}
sales_feb={
    'Customer':['A','B','C','D'],
    'Product':['Apple','Banana','Orange','Pineapple'],
    'Quantity':[10,30,60,80],
    'Price':[110,210,320,340]
}
sales_march={
    'Customer':['A','B','C','D'],
    'Product':['Apple','Banana','Orange','Pineapple'],
    'Quantity':[15,35,65,85],
    'Price':[80,211,430,230]
}
```

```python
df_1=pd.DataFrame(sales_jan)
df_2=pd.DataFrame(sales_feb)
df_3=pd.DataFrame(sales_march)
print(df_1,"\n",df_2,"\n",df_3)
```

```python
pd.concat([df_1,df_2,df_3]) 
```

## Merging 

```python
df_4=pd.DataFrame({
    'Customer':['E','F','G','H'],
    'Product':['Grapes','Mango','Watermelon','Kiwi'],
    'Quantity':[50,70,90,20],
    'Price':[150,250,350,450]
})
df_5=pd.DataFrame({
    'Customer_ID':[1112,1113,1114,1115],
    'Customer':['E','F','G','H'],
    'Location':['Delhi','Mumbai','Chennai','Kolkata'],
    'Age':[25,35,45,55]
})

pd.merge(df_4,df_5, on='Customer')
```

## Joining


```python
sales_info=pd.DataFrame({
    'Customer':['A','B','C','D'],
    'Product':['Apple','Banana','Orange','Pineapple'],
    'Quantity':[5,3,6,8],
    'Price':[100,200,300,400]
},index=[1,2,3,4])

store_info=pd.DataFrame({
    'Store_ID':[101,102,103,104],
    'Store_Name':['Store_A','Store_B','Store_C','Store_D'],
    'CoDe':['A','2','C','4']
},index=[0,1,2,3])
```

```python
sales_info.join(store_info)
```

```python
sales_info.join(store_info,how='outer')
```

##Shows all the indexes data regardless of matching situation.

```python
sales_info.join(store_info,how='inner')
```

##Shows data of ony matching indexes
