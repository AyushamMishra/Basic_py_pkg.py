## Bar plot

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
Dev_names=['Ashish','Rohit','Sagar','Ankit','Rakesh']
Salary=[45000,55000,60000,75000,80000]
Py_dev= [47000,50000,63400,63000,45000]
Js_dev=[42000,52000,58000,60000,52000]

x_indexes=np.arange(len(Dev_names))
width=0.25

plt.bar(x_indexes-width,Salary,color='orange',width=width,label='Avg Salary')
plt.bar(x_indexes,Py_dev,color='red',width=width,label='Python Dev')
plt.bar(x_indexes+width,Js_dev,color='green',width=width,label='JavaScript Dev')

plt.xlabel('Developer names')
plt.ylabel('Salary')
plt.grid(True)
plt.title('Developer salary')
plt.legend()
plt.xticks(ticks=x_indexes,labels=Dev_names)
plt.show()
````

## Histogram

A histogram is a type of plot in which:

   The x-axis is divided into intervals (called bins).

   The y-axis shows the frequency (count) of data points falling into each bin.

   Unlike a bar chart (which compares categories), a histogram shows the distribution of numerical data.

Uses of Histograms

1) Understand distribution  e.g., normally distributed, skewed, uniform.

2) Detect outliers  see unusual data points.

3) Measure spread & central tendency  visualize mean/median around data.

4) Compare datasets  see if two groups differ in distribution.

5) Data preprocessing in ML/AI  check balance of features, normalization needs.

Histogram = a graphical summary of data distribution

````python
import matplotlib.pyplot as plt
import numpy as np
````

````python
np.random.seed(36)
ages=np.random.randint(1,110,50)
ages
````

````python
median_age=np.median(ages)

counts,bin_edges,_=plt.hist(ages,bins=10,edgecolor='black',color='orange',alpha=0.7,histtype='barstacked',orientation='vertical',rwidth=0.9,density=False,
         cumulative=False,align='mid',label='Age groups',stacked=True,log=False,fill=True,linewidth=1.5,hatch='//')

print(counts,bin_edges)

plt.axvline(median_age,color='red',linestyle='dashed',label='Median age')

plt.xlabel('Ages')
plt.ylabel('No of people')
plt.title('Age Distribution')                                     ## histtype - used for different types of histogram (barstacked,step,stepfilled)
plt.legend()                                                      ## orientation - vertical/horizontal
plt.show()                                                        ## rwidth - relative width of bars
                                                                  ## density - if True, normalizes the histogram
                                                                  ## cumulative - if True, cumulative histogram
                                                                  ## align - alignment of bars (left,mid,right)
                                                                  ## stacked - if True, multiple data are stacked
                                                                  ## log - if True, uses log scale for y axis
                                                                  ## fill - if True, fills the bars with color

````

## Subplots

Subplots are power full tool that allow us to use creat many plots within a single figure 

````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
````

````python
Student_data={
    'Names':['Jhon','Alice','Bob','Ben','Srock'],
    'Science':[56,65,77,23,78],
    'Maths':[22,34,22,76,11],
    'English':[33,44,56,73,74]
}
df=pd.DataFrame(Student_data,index=['a','b','c','d','e'])
df
````

````python
fig, ax=plt.subplots(2,1,sharex=True,sharey=True)   ## 2 rows and 1 column 
ax[0].plot(df['Names'],df['English'],label='English marks',color='red')
ax[0].set_xlabel('Student_Names')
ax[0].set_ylabel('Eng_Marks')
ax[0].set_title('English')
ax[0].legend()
 ## ax[0].tick_params(axis='x',rotation=90) using a common x axis

ax[1].plot(df['Names'],df['Maths'],label='Maths marks')
ax[1].set_xlabel('Student_Names')
ax[1].set_ylabel('Math_Marks')
ax[1].set_title('Maths')
ax[1].legend()
plt.tight_layout()
plt.show()
````

## Subplot Methods and attributes

#  `plt.subplots()`  Basics

```python
fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 6))
```

* **`fig`**  the **Figure** object (the whole canvas).

* **`ax`**  the **Axes** object(s) (individual subplots inside the figure).

* `nrows`, `ncols`  number of subplot rows & columns.

* `figsize=(width, height)`  size of the whole figure.

---

#  Key Methods & Attributes

### 1. Figure (`fig`) methods/attributes

| Attribute/Method          | Description                                     |
| ------------------------- | ----------------------------------------------- |
| `figsize`                 | Size of the figure.                             |
| `fig.suptitle("Title")`   | Adds a **main title** for the whole figure.     |
| `fig.subplots_adjust()`   | Adjusts spacing (top, bottom, wspace, hspace).  |
| `fig.tight_layout()`      | Automatically adjusts spacing to avoid overlap. |
| `fig.colorbar()`          | Add colorbars for heatmaps/images.              |
| `fig.savefig("plot.png")` | Save the whole figure.                          |

---

### 2. Axes (`ax`) methods/attributes

| Attribute/Method                            | Description                     |
| ------------------------------------------- | ------------------------------- |
| `ax.plot()`, `ax.scatter()`                 | Plot inside a specific subplot. |
| `ax.set_title("Title")`                     | Title for that subplot.         |
| `ax.set_xlabel("X")` / `ax.set_ylabel("Y")` | Axis labels.                    |
| `ax.legend()`                               | Show legend inside subplot.     |
| `ax.grid(True)`                             | Add gridlines.                  |
| `ax.set_xlim()`, `ax.set_ylim()`            | Set axis limits.                |
| `ax.imshow()`                               | Show images in subplot.         |
| `ax.hist()`                                 | Histogram in subplot.           |
| `ax.text(x, y, "Label")`                    | Annotate text inside plot.      |

---

#  Arranging Subplots

### 1. Single row/column

```python
fig, ax = plt.subplots(1, 3, figsize=(12, 4))
ax[0].plot(x, y1)
ax[1].scatter(x, y2)
ax[2].hist(y3)
```

### 2. 2D grid (matrix style)

```python
fig, ax = plt.subplots(2, 2, figsize=(10, 8))
ax[0,0].plot(x, y)
ax[0,1].bar(categories, values)
ax[1,0].hist(data)
ax[1,1].imshow(image)
```

---

#  Useful Attributes

* `fig.axes`  list of all subplot axes in figure.
* `ax.get_xlim()` / `ax.get_ylim()`  current axis range.
* `ax.set_aspect("equal")`  fix aspect ratio.
* `ax.sharex`, `ax.sharey`  share axis across subplots.

Example:

```python
fig, ax = plt.subplots(2, 2, sharex=True, sharey=True)
```

---
 **Cheat sheet summary:**

* `fig` = whole canvas (titles, spacing, save).
* `ax` = each subplot (labels, plot, limits).
* Use `sharex/sharey` for linked plots.
* Use `tight_layout()` to avoid overlap.
* Use `subplots_adjust()` for fine-tuning.



## Specific usecase of various types of plots in matplotlib in ML

##  1. Line Plot (`plt.plot`)

**Use case in ML:**

* To show **training vs validation loss/accuracy curves** during model training (deep learning, regression, etc.).
* To visualize **learning rate schedules** in optimizers.

```python
plt.plot(history['loss'], label="Training Loss")
plt.plot(history['val_loss'], label="Validation Loss")
plt.legend()
```

---

##  2. Scatter Plot (`plt.scatter`)

**Use case in ML:**

* To visualize **relationships between two features**.
* To see **class separation** in 2D (good for clustering/ classification).
* Used for **decision boundary visualization** in simple models.

---

##  3. Histogram (`plt.hist`)

**Use case in ML:**

* To check **distribution of features** (are they normal, skewed, uniform?).
* To spot **outliers** and **imbalanced data classes**.
* To visualize **prediction probability distributions** (e.g., how confident your model is).

---

##  4. Bar Chart (`plt.bar`)

**Use case in ML:**

* To show **class frequencies** in imbalanced datasets.
* To compare **feature importance scores** from models like Random Forest or XGBoost.
* To compare **model metrics (accuracy, F1, precision, recall)** across models.

---

##  5. Box Plot / Violin Plot (`plt.boxplot`, `plt.violinplot`)

**Use case in ML:**

* To detect **outliers** in features.
* To compare **data distributions** across classes.
* To visualize **cross-validation results** for error spread.

---

##  6. Heatmap (`plt.imshow` + `seaborn.heatmap`)

(Matplotlib itself can plot with `imshow`, but usually used with Seaborn.)
**Use case in ML:**

* To show **correlation matrix of features** (feature selection).
* To visualize a **confusion matrix** for classification performance.
* To show **attention maps** or **feature activations** in deep learning.

---

##  7. Area Plot (`plt.fill_between`)

**Use case in ML:**

* To visualize **confidence intervals** around regression predictions.
* To highlight **variance across folds** in cross-validation.

---

##  8. Contour Plot (`plt.contour`, `plt.contourf`)

**Use case in ML:**

* To visualize **decision boundaries** of classifiers (SVM, logistic regression, etc.).
* To illustrate **loss surfaces** in optimization research.

---

##  9. Image Display (`plt.imshow`)

**Use case in ML:**

* To visualize **image datasets** (MNIST, CIFAR-10).
* To show **filters/kernels** learned by CNNs.
* To display **segmentation masks or predictions**.

---

##  10. Error Bars (`plt.errorbar`)

**Use case in ML:**

* To show **uncertainty in predictions**.
* To plot **mean  std** across multiple experiment runs.

---

 **Summary (Cheat Sheet):**

* **Line**  training curves, optimization.
* **Scatter**  class separation, feature relation.
* **Histogram**  distributions, imbalance, outliers.
* **Bar**  class counts, feature importance, metrics.
* **Box/Violin**  outlier detection, variance.
* **Heatmap**  correlations, confusion matrix.
* **Area**  uncertainty/confidence intervals.
* **Contour**  decision boundaries, loss landscapes.
* **Image**  datasets, CNN filters, segmentation.
* **Error bars**  uncertainty, experiment variation.
