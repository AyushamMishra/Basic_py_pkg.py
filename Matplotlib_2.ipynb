## Bar plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Dev_names=['Ashish','Rohit','Sagar','Ankit','Rakesh']
Salary=[45000,55000,60000,75000,80000]
Py_dev= [47000,50000,63400,63000,45000]
Js_dev=[42000,52000,58000,60000,52000]

x_indexes=np.arange(len(Dev_names))
width=0.25

plt.bar(x_indexes-width,Salary,color='orange',width=width,label='Avg Salary')
plt.bar(x_indexes,Py_dev,color='red',width=width,label='Python Dev')
plt.bar(x_indexes+width,Js_dev,color='green',width=width,label='JavaScript Dev')

plt.xlabel('Developer names')
plt.ylabel('Salary')
plt.grid(True)
plt.title('Developer salary')
plt.legend()
plt.xticks(ticks=x_indexes,labels=Dev_names)
plt.show()
```
