## Line plot

```python
import matplotlib.pyplot as plt
```

```python
months=range(1,13)
Agra_temp=[7.9,9.6,15.3,22.6,28.2,32.5,30.8,30.1,27.2,21.3,14.4,9.8]
Delhi_temp=[7.8,10.2,16.6,25.0,32.2,36.0,34.9,33.8,30.3,24.1,16.7,10.6]
Mumbai_temp=[24.1,25.6,27.2,29.3,30.6,32.2,32.8,32.8,31.9,30.3,27.8,25.4]
```

```python
plt.plot(months,Agra_temp)
plt.show()
```

```python
plt.plot(months,Delhi_temp)
plt.show()
```

```python
plt.plot(months,Mumbai_temp,marker='d')
plt.xlabel('Months')
plt.ylabel('Temperature')
plt.show()
```

```python
plt.plot(months,Agra_temp,label='Agra',color='r',marker='o')
plt.plot(months,Delhi_temp,label='Delhi',color='g',marker='*')
plt.plot(months,Mumbai_temp,label='Mumbai',color='b',marker='d')
plt.title('Monthly average temprature')
plt.xlabel('Months')
plt.ylabel('Tempratures')
plt.legend()
plt.show()
```
